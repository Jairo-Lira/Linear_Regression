```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
url = 'https://raw.githubusercontent.com/Jairo-Lira/Linear_Regression/main/cost_revenue_clean.csv?token=GHSAT0AAAAAAB6NXRHHKDYZAYTF5DA7YVDAY7BKP5Q'
data = pd.read_csv(url)
```

```python
data.describe()
```

```python
X = pd.DataFrame(data, columns = ['production_budget_usd'])
y = pd.DataFrame(data, columns = ['worldwide_gross_usd'])
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(X, y, alpha = 0.3)
plt.title('Fil costs vs Global revenue')
plt.xlabel('Production budget $')
plt.ylabel('Worldwide gross $')
plt.ylim(0, 3e9)
plt.xlim(0, 45e7)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

Slope Coefficient

```python
regression.coef_ #Theta_1
```

```python
regression.intercept_
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(X, y, alpha = 0.3)
plt.plot(X, regression.predict(X), color = 'red', linewidth = 4)
plt.title('Fil costs vs Global revenue')
plt.xlabel('Production budget $')
plt.ylabel('Worldwide gross $')
plt.ylim(0, 3e9)
plt.xlim(0, 45e7)
plt.show()
```

```python
regression.score(X, y)
```
